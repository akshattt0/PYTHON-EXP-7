```python
# Multiplication of two 3x3 matrices (important)

A=[[1,2,3],[1,2,1],[7,4,5]]
B=[[0,1,2],[2,4,7],[4,6,7]]
Result=[[0,0,0],[0,0,0],[0,0,0]]

num_col=len(A[0])
num_rows=len(B)

if num_col==num_rows:
  for i in range(3):
    for j in range(3):
      for k in range(3):
        Result[i][j] += A[i][k] * B[k][j]
  print('Product:')
  for row in Result:
    print(row)
else:
  print('Matrices cant be multiplied')




```

```python
# Python program to accept three digits and print all possible combination


d1 = input("Enter first digit: ")
d2 = input("Enter second digit: ")
d3 = input("Enter third digit: ")

digits = [d1, d2, d3]

print("All possible combinations are:")

for i in range(3):
    for j in range(3):
        for k in range(3):
            if i != j and j != k and i != k:
                print(digits[i] + digits[j] + digits[k])
```

```python
#Amstrong number

num=int(input("Enter the number:"))
total =0

for digit in str(num):
  total += int(digit)**len(str(num))

if total==num:
  print('Amstrong number')
else:
  print('Not Amstrong')
```
